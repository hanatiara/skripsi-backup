```python
import torch
import onnx
import torchvision
from onnx2tflite import onnx_converter
```

```python
# Load  PyTorch model
model = torchvision.models.resnet50(pretrained=True)
# Set  input shape of the model
input_shape = (1, 3, 224, 224)
# Export  PyTorch model to ONNX format
torch.onnx.export(model, torch.randn(input_shape), 'resnet50.onnx', opset_version=11)
```
