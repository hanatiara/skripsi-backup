```python
import os
from collections import Counter
```

```python
# Misalnya, struktur folder adalah "dataset/class_name/image.jpg"
dataset_dir = 'cleaned-dataset'
class_counts = Counter()

# Iterasi folder untuk menghitung jumlah gambar per kelas
for class_name in os.listdir(dataset_dir):
    class_path = os.path.join(dataset_dir, class_name)
    if os.path.isdir(class_path):
        class_counts[class_name] = len(os.listdir(class_path))

print("Jumlah gambar per kelas:", class_counts)
```

```python
import matplotlib.pyplot as plt

# Plot jumlah gambar per kelas
plt.figure(figsize=(10, 5))
plt.bar(class_counts.keys(), class_counts.values())
plt.xlabel('Kelas')
plt.ylabel('Jumlah Gambar')
plt.title('Distribusi Jumlah Gambar per Kelas')
plt.xticks(rotation=45)
plt.show()
```

```python
image_paths = []
for class_name in os.listdir(dataset_dir):
    class_path = os.path.join(dataset_dir, class_name)
    if os.path.isdir(class_path):
        image_paths.extend([os.path.join(class_path, img) for img in os.listdir(class_path)])

means = []
stds = []

# Proses setiap gambar
for image_path in image_paths:
    image = Image.open(image_path).convert("RGB")  # Konversi ke RGB
    image_np = np.array(image) / 255.0  # Normalisasi nilai piksel ke rentang [0,1]
    
    # Hitung rata-rata dan standar deviasi per gambar
    means.append(image_np.mean(axis=(0, 1)))  # Mean per channel (R, G, B)
    stds.append(image_np.std(axis=(0, 1)))    # Std per channel (R, G, B)

# Hitung mean dan standard deviation keseluruhan dataset
mean = np.mean(means, axis=0)
std = np.mean(stds, axis=0)

print("Mean piksel dataset:", mean)
print("Standard Deviation piksel dataset:", std)
```

```python
import matplotlib.pyplot as plt
import random
from PIL import Image
import os
import math

# Asumsi dataset_dir dan class_counts sudah didefinisikan

num_classes = len(class_counts)
grid_size = math.ceil(num_classes ** 0.5)  # Menentukan ukuran grid yang mendekati akar kuadrat dari jumlah kelas

plt.figure(figsize=(12, 12))

for i, class_name in enumerate(class_counts.keys()):
    class_path = os.path.join(dataset_dir, class_name)
    sample_image = random.choice(os.listdir(class_path))
    image = Image.open(os.path.join(class_path, sample_image))

    plt.subplot(grid_size, grid_size, i + 1)
    plt.imshow(image)
    plt.title(class_name)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
sizes = []

for image_path in image_paths:
    image = Image.open(image_path)
    sizes.append(image.size)

# Melihat distribusi ukuran gambar
widths, heights = zip(*sizes)
print("Rata-rata ukuran gambar: Lebar =", np.mean(widths), ", Tinggi =", np.mean(heights))
```

```python
import hashlib

hashes = set()
duplicates = []

for image_path in image_paths:
    with open(image_path, 'rb') as f:
        file_hash = hashlib.md5(f.read()).hexdigest()
    if file_hash in hashes:
        duplicates.append(image_path)
    else:
        hashes.add(file_hash)

print("Jumlah gambar duplikat:", len(duplicates))
```
